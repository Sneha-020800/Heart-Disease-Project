### Chest Pain Types

```python
#Chest Pain Type (0–3)

0 → Typical Angina - Chest pain related to decreased blood supply to the heart (classic symptom of coronary artery disease).

1 → Atypical Angina - Chest pain not directly related to exertion, can mimic heart pain but may also be from non-cardiac causes.
    
2 → Non-Anginal Pain - Chest pain not due to the heart (e.g., gastric, musculoskeletal).

3 → Asymptomatic - No chest pain, but heart disease may still be present (often detected in routine check-ups or ECG results).
```

### Resting Blood Pressure Distribution Overview

```python
df.columns
```

```python
normal = (df['Resting Blood Pressure (mmHg)'] <= 120).sum()
at_risk = (df['Resting Blood Pressure (mmHg)'] > 120).sum()
total = len(df)

print(f"Normal (≤120 mmHg): {normal}")
print(f"At Risk (>120 mmHg): {at_risk}")
print(f"Percentage at risk: {at_risk/total*100:.2f}%")

ax = sns.histplot(df['Resting Blood Pressure (mmHg)'], bins=20)
for p in ax.patches:
    ax.text(p.get_x() + p.get_width()/2, p.get_height(), int(p.get_height()),
            ha='center', va='bottom')

# Vertical line at threshold 120
plt.axvline(120, color='red', linestyle='--', label='Normal threshold (120 mmHg)')

plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.show()
```

*Using a threshold of 120 mmHg, we find 838 patients (27.3%) are normal (≤120) and 2231 patients (72.7%) are at risk (>120). This shows that about three-quarters of the population has elevated resting blood pressure, a strong risk factor for heart disease. The red dashed line at 120 mmHg on the histogram marks this cutoff.*

### Cholestrol Vs Heart Disease

```python
plt.figure(figsize=(14,7))

# Histogram with heart disease hue
ax = sns.histplot(data=df, x='Cholesterol (mg/dl)', hue='heart_disease', kde=True, bins=20)

# 1. Add numbers on each bar
for p in ax.patches:
    height = p.get_height()
    if height > 0:
        ax.annotate(f'{int(height)}',
                    (p.get_x() + p.get_width() / 2., height),
                    ha='center', va='bottom', fontsize=15, color='black')

# 2. Add vertical cutoffs
plt.axvline(200, color='blue', linestyle='--', label='Normal/Borderline cutoff', linewidth=2.5)
plt.axvline(240, color='red', linestyle='--', label='High Risk cutoff', linewidth=2.5)

# 4. Calculate and print totals
normal = df[df['Cholesterol (mg/dl)'] < 200].shape[0]
borderline = df[(df['Cholesterol (mg/dl)'] >= 200) & (df['Cholesterol (mg/dl)'] < 240)].shape[0]
high = df[df['Cholesterol (mg/dl)'] >= 240].shape[0]

print(f"Normal (<200): {normal}")
print(f"Borderline (200-239): {borderline}")
print(f"high (>=240): {high}")

# 6. Move legend outside
ax.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

*Out of the total patients analyzed:*

*649 (≈21%) have cholesterol levels in the normal range (<200 mg/dl).*

*258 (≈8%) fall in the borderline range (200–239 mg/dl).*

*A very large majority, 2162 (≈71%), have high cholesterol (≥240 mg/dl), placing them at a significantly elevated risk.*

*This distribution clearly indicates that high cholesterol is strongly prevalent in this group and may be a major contributing factor to heart disease.*

### Analysis of BMI, Heart Rate, Blood Sugar and Smoking on Heart Disease

```python
df.columns
```

```python
fig, axs = plt.subplots(2, 2, figsize=(12, 8))

# Plot 1: Body Mass Index Categories (Bar Chart)
df['BMI Category'] = pd.cut(
    df['Body Mass Index'],
    bins=[0, 18.5, 24.9, 29.9, 100],
    labels=['Underweight', 'Normal', 'Overweight', 'Obese']
)

ax=sns.countplot(x='BMI Category',hue='BMI Category', data=df, ax=axs[0, 0], palette="Set2",order=['Underweight', 'Normal', 'Overweight', 'Obese'])
axs[0, 0].set_title("Body Mass Index Categories")
axs[0, 0].set_ylabel("Count")
for container in ax.containers:
    ax.bar_label(container, label_type='edge')

# Plot 2: Max Heart Rate vs Heart Disease (Bar Chart)
hr=df.groupby('heart_disease')['Max Heart Rate Achieved'].mean().plot(
    kind='bar', ax=axs[0, 1], color=['lightblue','lightpink']
)
axs[0, 1].set_title("Max Heart Rate vs Heart Disease")
axs[0, 1].set_xticklabels(['No Disease', 'Disease'], rotation=0)
for container in axs[0, 1].containers:
    axs[0, 1].bar_label(container, fmt='%.1f')   # one decimal place

# Plot 3: Fasting Blood Sugar vs Heart Disease (Stacked Bar)
crosstab = pd.crosstab(
    df['heart_disease'],
    df['Fasting Blood Sugar > 120 mg/dl (1 = Yes, 0 = No)']
)
crosstab.plot(
    kind='bar', stacked=True, ax=axs[1, 0],
    color=['lightblue', 'orange']
)
axs[1, 0].set_title("Fasting Blood Sugar vs Heart Disease")
axs[1, 0].set_xticklabels(['No Disease', 'Disease'], rotation=0)
axs[1, 0].set_ylabel("Count")
axs[1, 0].legend(['≤120 mg/dl', '>120 mg/dl'])
for container in axs[1, 0].containers:
    axs[1, 0].bar_label(container, label_type="center")  # labels inside the bars

# Plot 4: Smoking Status Distribution (Donut Chart)
smoke_counts = df['Smoking (1 = Yes, 0 = No)'].value_counts().sort_index()
axs[1, 1].pie(
    smoke_counts, labels=['Non-Smoker', 'Smoker'],
    autopct='%1.1f%%', colors=['#8dd3c7','#fb8072'],
    startangle=90, wedgeprops=dict(width=0.4)
)
axs[1, 1].set_title("Smoking Status Distribution")

plt.tight_layout()
plt.show()
```

BMI Distribution: Most individuals fall in the obese category (1210), followed by normal (783) and overweight (635),while only a small portion are 
underweight (441). This indicates a higher prevalence of obesity in the dataset.

Max Heart Rate vs Heart Disease: The average maximum heart rate is almost the same for both groups (135.1 in non-diseased vs 135.5 in diseased). 
    Hence, maximum heart rate does not show a strong difference between people with and without heart disease.

Fasting Blood Sugar vs Heart Disease: Among individuals without heart disease, the majority (1574) had blood sugar ≤120 mg/dl, 
while 256 had levels >120 mg/dl. Similarly, in those with heart disease, most (1048) had ≤120 mg/dl, but 191 had elevated sugar (>120 mg/dl). 
This suggests that while higher sugar levels are more frequent in heart disease cases proportionally, most people (both groups) still fall under
the normal sugar range.

Smoking Status Distribution: Around 65.1% are non-smokers and 34.9% are smokers, showing that about one-third of the population smokes, which is a 
notable risk factor for cardiovascular health.

Together, these insights suggest obesity and smoking are major concerns, while max heart rate is not a strong differentiator. 
    High fasting blood sugar shows some association with heart disease but is not the dominant factor since most cases still have normal sugar levels.
